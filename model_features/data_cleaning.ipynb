```python
import os
#Data handling
import pandas as pd
import numpy as np
import datetime
#Data Visualization
import matplotlib as mpl
import matplotlib.pyplot as plt
# use the inline backend to generate the plots within the browser
%matplotlib inline 
```

```python
#Loading datasets

df_beauty=pd.read_json('../data/ref/All_Beauty_5.json', lines=True)
df_appliances=pd.read_json('../data/ref/Appliances_5.json', lines=True)
meta_data=pd.read_json('../data/ref/meta_data.json', lines=True)
```

# Meta Data cleaning

```python
meta_data.head()
```

```python
#Converting the list for images, rank and description
meta_data['description']=meta_data['description'].apply(lambda x: x[0] if isinstance(x,list) else x)
meta_data['image']=meta_data['image'].apply(lambda x: x[0] if isinstance(x,list) else x)
meta_data['rank']=meta_data['rank'].apply(lambda x: x[0] if isinstance(x,list) else x)
```

```python
meta_data.head()
```

```python
#Convert image into a categorical indicating variable
meta_data['image']=meta_data['image'].apply(lambda x: 1 if isinstance(x,str) and 'https' in x else 0)
meta_data.head()
```

```python
#Clean rank
meta_data['rank']=meta_data['rank'].apply(lambda x: x.split('in',1)[0] if isinstance(x,str) else 0)
meta_data.head()
```

```python
for crc in ['>','#',',']:
    meta_data['rank']=meta_data['rank'].apply(lambda x:x.replace(crc,'') if isinstance(x,str) else 0)
meta_data.head()
```

```python
#Filtering on the columns we want to use
meta_data=meta_data[['asin','date','title','description','image','main_cat','brand','rank']]
```

```python
#Creating year column
meta_data['product_year']=meta_data['date'].dt.year
meta_data['product_year']=meta_data['product_year'].fillna(-1)
meta_data['product_year']=meta_data['product_year'].astype(int)
```

```python
meta_data['product_year'].value_counts(dropna=False)
```

```python
meta_data.head()
```

```python
meta_data.shape
```

```python
#Dropping duplicates
meta_data.drop_duplicates(subset=['asin','title'], inplace=True)
meta_data.shape
```

# Appliances Reviews Dataset Cleaning

```python
df_appliances.head()
```

```python
#Filling N/A in vote column
df_appliances['vote']=df_appliances['vote'].fillna(0)
df_appliances['vote']=df_appliances['vote'].apply(lambda x: x.replace(',','') if isinstance(x,str) else x)
```

```python
df_appliances['review_date']=df_appliances['unixReviewTime'].apply(lambda x:datetime.datetime.utcfromtimestamp(x).strftime('%Y-%m-%d'))
```

```python
df_appliances.head()
```

```python
df_appliances['review_date']=df_appliances['review_date'].astype('datetime64')
df_appliances['review_date'].dtype
```

```python
df_appliances['review_date']
```

```python
#Filtering and reordering columns

df_appliances=df_appliances[['asin','review_date','reviewerID','reviewerName','summary','overall','reviewText','vote']]
df_appliances.head()
```

# Joining the data

```python
df=df_appliances.merge(meta_data, how='left', on='asin')
df.head()
```
