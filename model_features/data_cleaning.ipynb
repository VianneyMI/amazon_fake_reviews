```python
import os
#Data handling
import pandas as pd
import numpy as np
import datetime
#Data Visualization
import matplotlib as mpl
import matplotlib.pyplot as plt
# use the inline backend to generate the plots within the browser
%matplotlib inline 
```

```python
#Loading datasets

#df_beauty=pd.read_json('../data/ref/All_Beauty_5.json', lines=True)
df_appliances=pd.read_json('../data/ref/Appliances_5.json', lines=True)
meta_appliances=pd.read_json('../data/ref/meta_Appliances.json', lines=True)
```

# Meta Data cleaning

```python
meta_appliances.head()
```

```python
#Converting the list for images, rank, description and features,
meta_appliances['description']=meta_appliances['description'].apply(lambda x: x[0] if isinstance(x,list) and x else 'N/A')
meta_appliances['image']=meta_appliances['image'].apply(lambda x: x[0] if isinstance(x,list) and x else 'N/A')
meta_appliances['rank']=meta_appliances['rank'].apply(lambda x: x[0] if isinstance(x,list) and x else 'N/A')
meta_appliances['feature']=meta_appliances['feature'].apply(lambda x: x[0] if isinstance(x,list) and x else 'N/A')
```

```python
meta_appliances.head()
```

```python
#Convert image into a categorical indicating variable
meta_appliances['image']=meta_appliances['image'].apply(lambda x: 1 if isinstance(x,str) and 'https' in x else 0)
meta_appliances.head()
```

```python
#Clean rank
meta_appliances['rank']=meta_appliances['rank'].apply(lambda x: x.split('in',1)[0] if isinstance(x,str) else 0)
meta_appliances.head()
```

```python
for crc in ['>','#',',']:
    meta_appliances['rank']=meta_appliances['rank'].apply(lambda x:x.replace(crc,'') if isinstance(x,str) else -1)
meta_appliances.head()
```

```python
#Filtering on the columns we want to use
meta_appliances=meta_appliances[['asin','date','title','description','feature','image','main_cat','brand','rank']]
```

```python
meta_appliances['date']=pd.to_datetime(meta_appliances['date'], errors='coerce')
```

```python
#Creating year column
meta_appliances['productYear']=meta_appliances['date'].dt.year
meta_appliances['productYear']=meta_appliances['productYear'].fillna(-1)
meta_appliances['productYear']=meta_appliances['productYear'].astype(int)
```

```python
#Creating month column
meta_appliances['productMonth']=meta_appliances['date'].dt.month
meta_appliances['productMonth']=meta_appliances['productMonth'].fillna(-1)
meta_appliances['productMonth']=meta_appliances['productMonth'].astype(int)
```

```python
#Creating day column
meta_appliances['productDay']=meta_appliances['date'].dt.day
meta_appliances['productDay']=meta_appliances['productDay'].fillna(-1)
meta_appliances['productDay']=meta_appliances['productDay'].astype(int)
```

```python
meta_appliances['productYear'].value_counts(dropna=False)
```

```python
meta_appliances.head()
```

```python
meta_appliances.shape
```

```python
#Dropping duplicates
meta_appliances.drop_duplicates(subset=['asin','title'], inplace=True)
meta_appliances.shape
```

```python
meta_appliances.columns
```

```python
meta_appliances.rename({
    'date':'productDate',
    'title':'productTitle',
    'description':'productDescription',
    'feature':'productFeatures',
    'image':'productHasImage',
    'main_cat':'productCategory',
    'brand':'productBrand',
    'rank':'productRank'
}, axis=1, inplace=True)
meta_appliances.head()
```

# Appliances Reviews Dataset Cleaning

```python
df_appliances.head()
```

```python
df_appliances['verified'].value_counts()
```

```python
df_appliances['image']=df_appliances['image'].apply(lambda x: x[0] if isinstance(x,list) and x else 'N/A')
```

```python
df_appliances['image']
```

```python
#Convert image into a categorical indicating variable
df_appliances['image']=df_appliances['image'].apply(lambda x: 1 if isinstance(x,str) and 'https' in x else 0)
df_appliances.head()
```

```python
df_appliances['image'].value_counts()
```

```python
#Filling N/A in vote column
df_appliances['vote']=df_appliances['vote'].fillna(0)
df_appliances['vote']=df_appliances['vote'].apply(lambda x: x.replace(',','') if isinstance(x,str) else x)
```

```python
df_appliances['reviewDate']=df_appliances['unixReviewTime'].apply(lambda x:datetime.datetime.utcfromtimestamp(x).strftime('%Y-%m-%d'))
```

```python
df_appliances.head()
```

```python
df_appliances['reviewDate']=df_appliances['reviewDate'].astype('datetime64')
df_appliances['reviewDate'].dtype
```

```python
df_appliances['reviewDate']
```

```python
#Filtering and reordering columns

df_appliances=df_appliances[['asin','reviewDate','reviewerID','reviewerName','summary','overall','reviewText','vote','image','verified']]
df_appliances.head()
```

```python
#Creating year column
df_appliances['reviewYear']=df_appliances['reviewDate'].dt.year
df_appliances['reviewYear']=df_appliances['reviewYear'].fillna(-1)
df_appliances['reviewYear']=df_appliances['reviewYear'].astype(int)
```

```python
#Creating month column
df_appliances['reviewMonth']=df_appliances['reviewDate'].dt.month
df_appliances['reviewMonth']=df_appliances['reviewMonth'].fillna(-1)
df_appliances['reviewMonth']=df_appliances['reviewMonth'].astype(int)
```

```python
#Creating day column
df_appliances['reviewDay']=df_appliances['reviewDate'].dt.day
df_appliances['reviewDay']=df_appliances['reviewDay'].fillna(-1)
df_appliances['reviewDay']=df_appliances['reviewDay'].astype(int)
```

```python
df_appliances.rename({
    'summary':'reviewTitle',
    'overall':'productGradeByReview',
    'vote':'reviewUpvotes',
    'image':'reviewHasImage',
    'verified':'reviewIsVerified'
}, axis=1, inplace=True)
df_appliances.head()
```

```python
df_appliances['reviewIsVerified'].dtype
```

```python
df_appliances['reviewIsVerified']=df_appliances['reviewIsVerified'].apply(lambda x: 1 if x else 0)
df_appliances['reviewIsVerified'].value_counts()
```
