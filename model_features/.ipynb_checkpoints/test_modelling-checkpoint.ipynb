```python
#General purpose
import os
#Data handling
import pandas as pd
import numpy as np
import datetime
#Data Visualization
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline 
#Modelling 
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from langdetect import detect #to filter non-english reviews.
import sklearn
```

```python
#Load dataset
df=pd.read_csv('../data/exports/test_appliances.csv')
#df.drop('Unnamed: 0',axis=1, inplace=True)
df.head()
```

```python
# Appending language
df['reviewLang']=df['reviewText'].apply(lambda x : detect(x)) 
```

```python
df.head()
```

```python
df.shape
```

```python
#Filtering on english reviews
dff=df[df['reviewLang']=='en']
dff.head()
```

```python
dff.shape
```

```python
nltk.download('stopwords')
stop_words=stopwords.words('english')
```

```python
reviews=dff['reviewText']
REVIEWS=reviews.tolist()
```

```python
type(REVIEWS)
```

# Tokenization

```python
nltk.download('punkt')
tokens=[word_tokenize(review) for review in REVIEWS]
```

```python
tokens[0]
```

```python
len(tokens)
```

```python
tokens2=[[]]
for i,token in enumerate(tokens):
    tokens2[i]=[word for word in token if word not in stop_words]
```
